## Building Linear Regression, ARIMA, and ARIMAX models

```python
import pandas as pd
import numpy as np
import statsmodels.tsa.stattools as sm
import statsmodels.tsa.arima_model as ARIMA
from statsmodels.tsa.stattools import adfuller
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from statsmodels.graphics.tsaplots import plot_acf
from pandas import tseries
from matplotlib import pyplot
from statsmodels.graphics.tsaplots import plot_pacf
from sklearn import linear_model
from sklearn.metrics import mean_squared_error
```

```python
# Import/clean dataframes

X_train = pd.read_csv('X_pca_train.csv', index_col='time')
X_test = pd.read_csv('X_pca_test.csv', index_col='time')
Y_train = pd.read_csv('Y_train.csv', index_col='time')
Y_test = pd.read_csv('Y_test.csv', index_col='time')
X = pd.concat([X_train, X_test])
Y = pd.concat([Y_train, Y_test])

X_train.index = pd.to_datetime(X_train.index)
X_test.index = pd.to_datetime(X_test.index)
Y_train.index = pd.to_datetime(Y_train.index)
Y_test.index = pd.to_datetime(Y_test.index)
```

```python
# Linear regression

# Using StatsModels OLS for model summary
linreg = sm.OLS(endog=Y,exog=X)
linreg_results = linreg.fit()
print(linreg_results.summary())

# Plot the fitted values
plt.plot(np.arange(len(linreg_results.fittedvalues)), linreg_results.fittedvalues, color='purple')
plt.plot(np.arange(len(Y)), Y, color='green')
plt.xlim((0, 1000))
plt.title('Linear Regression Fitted Values')
plt.xlabel('Time')
plt.ylabel('Price')
plt.show()

# Using sklearn LinearRegression() for out of sample forcasting
linreg_sklearn = linear_model.LinearRegression()
linreg_sklearn.fit(X_train, Y_train)
linreg_pred = linreg_sklearn.predict(X_test)
mse = np.sqrt(mean_squared_error(Y_test, linreg_pred))
print('R-square:', linreg_sklearn.score(X_test, Y_test))
print('rms:', mse)

# Plot the prediction
plt.plot(np.arange(len(linreg_pred)), linreg_pred, color='green', label='Prediction')
plt.plot(np.arange(len(Y_test)), Y_test, color='purple', label='Actual')
#plt.xlim((0, 50))
plt.title('Linear Regression prediction')
plt.xlabel('Time')
plt.ylabel('Price')
plt.legend()
plt.show()
```

```python
# Plotting ACF and PACF

# ACF
plot_acf(Y)
plt.title('ACF', loc='center')
plt.xlim((0, 50))
plt.ylabel('ACF')
plt.xlabel('Lag(s)')
plt.show()

# PACF
plot_pacf(Y, lags=10)
plt.title('PACF', loc='center')
plt.ylabel('PACF')
plt.xlabel('Lag(s)')
plt.show()
```

```python
# ARIMAX

# Building AR 2 with exogeneous variables
arima_d = ARIMA.ARIMA(endog=Y_train['XEM'], exog=X_train, order=[2,0,0])
arima_results_d = arima_d.fit()
print(arima_results_d.summary())

# Plotting the fitted values
plt.plot(np.arange(len(arima_results_d.fittedvalues)), arima_results_d.fittedvalues, color='purple')
plt.plot(np.arange(len(Y_train)), Y_train, color='green')
plt.xlim((0, 400))
plt.title('ARIMAX Fitted Values')
plt.xlabel('Time')
plt.ylabel('Price')
plt.show()

# Out-of-sample prediction
exog_d = X_test
arima_results_ofs_d = arima_results_d.predict(exog=exog_d, start=exog_d.index[0], end=exog_d.index[-1])
plt.plot(np.arange(len(Y_test)), Y_test, color='purple', label='Actual')
plt.plot(np.arange(len(arima_results_ofs_d)), arima_results_ofs_d, color='green', label='Prediction')
#plt.xlim((0, 50))
plt.title('ARIMAX Out-of-Sample Prediction')
plt.xlabel('Time')
plt.ylabel('Price')
plt.legend()
plt.show()
```

```python
# Building AR 2 model with no exogeneous variables

arima_1 = ARIMA.ARIMA(Y_train, order=(2,0,0))
arima_results_1 = arima_1.fit(disp=0)
print(arima_results_1.summary())
plt.plot(np.arange(len(arima_results_1.fittedvalues)), arima_results_1.fittedvalues, color='purple')
plt.plot(np.arange(len(Y)), Y, color='green')
plt.xlim((0, 400))
plt.title('ARIMA Fitted Values')
plt.xlabel('Time')
plt.ylabel('Price')
plt.show()

# Out-of-sample prediction
arima_results_ofs_1 = arima_results_1.predict(start=Y_test.index[0], end=Y_test.index[-1])
#arima_results_ofs_1 = arima_results_1.forecast(steps=600)[0]


plt.plot(np.arange(len(Y_test)), Y_test, color='purple', label='Actual')
plt.plot(np.arange(len(arima_results_ofs_1)), arima_results_ofs_1, color='green', label='Prediction')
#plt.xlim((0, 100))
plt.title('ARIMA Out-of-Sample Prediction')
plt.xlabel('Time')
plt.ylabel('Price')
plt.legend()
plt.show()

plt.plot(np.arange(len(arima_results_ofs_1)), arima_results_ofs_1, color='green', label='Prediction')
#plt.xlim((0, 100))
plt.title('ARIMA Out-of-Sample Prediction')
plt.xlabel('Time')
plt.ylabel('Price')
plt.legend()
plt.show()

plt.plot(np.arange(len(arima_results_ofs_1)), arima_results_ofs_1, color='green', label='Prediction')
plt.xlim((0, 20))
plt.title('ARIMA Out-of-Sample Prediction')
plt.xlabel('Time')
plt.ylabel('Price')
plt.legend()
plt.show()
```

```python
# statsmodels result for ARIMA with no exog was strange
# trying ARIMA using pyflux package

import pyflux as pf
model = pf.ARIMA(data=Y_train, ar=1, ma=0, target='XEM', family=pf.Normal())
x = model.fit("MLE")
x.summary()
pred_arima = model.predict(h=20)

#plt.plot(np.arange(len(y_test_d)), y_test_d, color='purple', label='Actual')
plt.plot(np.arange(len(pred_arima)), pred_arima, color='green', label='Predicted Price')
#plt.xlim(0,50)
plt.xlabel('Time')
plt.ylabel('Price')
plt.legend()
plt.show()
```
