## Building ARIMAX base model 

```python
import pandas as pd
import numpy as np
import statsmodels.tsa.stattools as sm
import statsmodels.tsa.arima_model as ARIMA
from statsmodels.tsa.stattools import adfuller
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from statsmodels.graphics.tsaplots import plot_acf
from pandas import tseries
from matplotlib import pyplot
from statsmodels.graphics.tsaplots import plot_pacf

from sklearn import linear_model
from sklearn.metrics import mean_squared_error
```

```python
# Import dataframe and create combination pairs

df = pd.read_csv('cap1_df.csv')

variables = {}
for i in df.columns: 
    if i != 'XEM':
        if i != 'time':
            variables[i] = df.loc[:, ('XEM', i)]
```

```python
# Performing Granger Causality Test

lag = 20
cor = {}
for i in variables.keys():
    cor[i] = sm.grangercausalitytests(variables[i], lag)  
```

```python
# Granger Causality test results

for key, values in cor.items():
    print('\n')
    print('\033[1m' + key + '\033[0m')
    for i in range(1, lag+1):
        print('lag', i, '=', values[i][0]['lrtest'][1])
```

```python
# Performing Granger Test on significan variables with maxlag = 50

#listing out statistically significant variables that Granger cause nem prices (based on the results above)
sig = ['rd_nem_g', 'gg_gpu', 'gg_nem_w']

rd_nem_g = sm.grangercausalitytests(variables['rd_nem_g'], 50)
gg_gpu = sm.grangercausalitytests(variables['gg_gpu'], 50)
gg_nem_w = sm.grangercausalitytests(variables['gg_nem_w'], 50)
```

```python
# Significant variables Grenger Test results

for i in sig:
    print('\n')
    print('\033[1m' + i + '\033[0m')
    for key, values in eval(i).items():
        print('lag', key, '=', values[0]['lrtest'][1])   
```

```python
# Construct a df with only significant variables

df = pd.read_csv('cap1_df.csv').set_index('time')
df_sig = pd.DataFrame(index=df.index)
df_sig['XEM'] = df['XEM']
df_sig['IOT'] = df['IOT']
df_sig['rd_nem_g'] = df['rd_nem_g']
df_sig['gg_gpu'] = df['gg_gpu']
df_sig['gg_nem_w'] = df['gg_nem_w']
df_sig = df_sig.loc['2017-08-29 00:00:00':]

# Feature scaling
sc_x = StandardScaler()
df_sig['XEM'] = sc_x.fit_transform(df_sig['XEM'].values.reshape(-1,1))
df_sig['IOT'] = sc_x.fit_transform(df_sig['IOT'].values.reshape(-1,1))
df_sig['rd_nem_g'] = sc_x.fit_transform(df_sig['rd_nem_g'].values.reshape(-1,1))
df_sig['gg_gpu'] = sc_x.fit_transform(df_sig['gg_gpu'].values.reshape(-1,1))
df_sig['gg_nem_w'] = sc_x.fit_transform(df_sig['gg_nem_w'].values.reshape(-1,1))
```

```python
# Performing Dickey-Fuller stationary test

for i in df_sig.columns: 
    x = df_sig[i].values
    result = adfuller(x)
    print('\033[1m' + i + '\033[0m')
    print('ADF Statistic:', result[0])
    print('p-value:', result[1])
```

```python
# Visualize the unstationary nature of data

plt.plot(np.arange(len(df_sig)), df_sig['XEM'], color='red')
plt.plot(np.arange(len(df_sig)), df_sig['IOT'], color='orange')
plt.plot(np.arange(len(df_sig)), df_sig['rd_nem_g'], color='blue')
plt.plot(np.arange(len(df_sig)), df_sig['gg_gpu'], color='green')
plt.plot(np.arange(len(df_sig)), df_sig['gg_nem_w'], color='purple')
plt.show()
```

```python
# Perfrom differencing to stationalize the series

# Creat difference function, with default value of lag 24
def difference(dataset, interval=24):
    diff = list()
    for i in range(interval, len(dataset)):
        value = dataset[i] - dataset[i - interval]
        diff.append(value)
    return diff

# Differencing the dataset
inter_d = 24
df_diff = pd.DataFrame(index=df_sig.index)
for i in df_sig.columns:
    if i != 'XEM':
        if i!= 'IOT':
            data = difference(df_sig[i], inter_d)
            data = pd.Series(np.append(np.repeat(np.nan, inter_d), data), index=df_sig.index, name=i)
            df_diff = pd.concat([df_diff, data], ignore_index=False, axis=1)
inter_h = 1
for i in ['XEM', 'IOT']:
        data = difference(df_sig[i], inter_h)
        data = pd.Series(np.append(np.repeat(np.nan, inter_h), data), index=df_sig.index, name=i)
        df_diff = pd.concat([df_diff, data], ignore_index=False, axis=1)
```

```python
# Re-perform Dickey-fuller test to test the differenced series for stationality

for i in df_diff.columns: 
    x = df_diff[i].dropna().values
    result = adfuller(x)
    print('\033[1m' + i + '\033[0m')
    print('ADF Statistic:', result[0])
    print('p-value:', result[1])
```

```python
# Visualize the stationarized data

plt.plot(np.arange(len(df_diff['XEM'])), df_diff['XEM'], color='red')
plt.plot(np.arange(len(df_diff['IOT'])), df_diff['IOT'], color='orange')
plt.plot(np.arange(len(df_diff['rd_nem_g'])), df_diff['rd_nem_g'], color='blue')
plt.plot(np.arange(len(df_diff['gg_gpu'])), df_diff['gg_gpu'], color='green')
plt.plot(np.arange(len(df_diff['gg_nem_w'])), df_diff['gg_nem_w'], color='purple')
plt.show()
```

```python
# Use IV 24 hour lag values for each DV time instance then make test/train sets

# Substitute XEM 24 hour lag values onto each IV time instance
df_pred = pd.DataFrame()
count = 0
for column in df_diff.columns:
    if column != 'XEM':
        df_pred[column] = df_diff.iloc[inter_d:, count].values
    count = count + 1

DV = df_diff['XEM'].values
DV = DV[inter_d * 2:]
DV = np.append(DV, np.repeat(np.nan, inter_d))
df_pred['XEM'] = DV
df_pred = df_pred.dropna()

# Test/train split
Xtrain, Xtest, Ytrain, Ytest = train_test_split(df_pred[['rd_nem_g', 'gg_gpu', 'gg_nem_w', 'IOT']], df_pred['XEM'], test_size = 0.3)
```

```python
# Plotting ACF and PACF

plot_acf(df_diff.loc['2017-08-29 01:00:00':,'XEM'])
plt.title('ACF', loc='center')
plt.xlim((0, 10))
plt.ylabel('ACF')
plt.xlabel('Lag(s)')
plt.show()

plot_pacf(df_diff.loc['2017-08-29 01:00:00':,'XEM'], lags=10)
plt.title('PACF', loc='center')
plt.ylabel('PACF')
plt.xlabel('Lag(s)')
plt.show()
```

```python
# ARIMA with 4 exogeneous variables

df_arima = pd.DataFrame(index=df_diff.index[inter_d:-inter_d])
for i in ['rd_nem_g', 'gg_gpu', 'gg_nem_w', 'IOT', 'XEM']:
    df_arima[i] = df_pred[i].values


model3 = ARIMA.ARIMA(endog=df_arima['XEM'], exog=df_arima[['rd_nem_g', 'gg_gpu', 'gg_nem_w', 'IOT']], order=[2,0,0])
results3 = model3.fit().predict()
plt.plot(np.arange(len(results3)), results3, color='purple')
plt.plot(np.arange(len(df_arima['XEM'])), df_arima['XEM'], color='green')
plt.xlim((100, 200))
plt.show()
```

```python
# Linear regression

model2 = sm.OLS(endog=df_pred['XEM'],exog=df_pred[['rd_nem_g', 'gg_gpu', 'gg_nem_w', 'IOT']])
results2 = model2.fit()
print(results2.summary())
```

```python
reg = linear_model.LinearRegression()
reg.fit(Xtrain, Ytrain)
pred = reg.predict(Xtest)
mse = np.sqrt(mean_squared_error(Ytest, pred))
print('R-square:', reg.score(Xtest, Ytest))
print('rms:', mse)

#scale = StandardScaler()
#predict = scale.fit_transform(pred)
#real = scale.fit_transform(Ytest)

plt.plot(np.arange(len(Ytest)), pred, color='purple')
plt.plot(np.arange(len(Ytest)), Ytest, color='green')
plt.xlim((0, 100))
plt.show()

#plt.plot(np.arange(len(Ytest)), real, color='green')
#plt.xlim((0, 100))
#plt.show()
```
